## ✅ 보조 기능

- `legend()` : 범례 표시  
- `xlabel()`, `ylabel()`, `title()` : 축/제목 설정  
- `xlim()`, `ylim()` : 축 범위 조절  
- `annotate()` : 그래프 내 텍스트 표시  
- `colorbar()` : 컬러맵 범례 추가  
- `grid()` : 격자 표시
